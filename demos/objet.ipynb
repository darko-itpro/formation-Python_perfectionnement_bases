```python
from datetime import datetime

class Training:
    def __init__(self, subject: str, duration: int, seats: int = 12):
        self.subject: str = subject
        self.duration = duration
        self.seats = seats
        self._students:list[tuple[str, datetime]] = []

    @property
    def remaining_seats(self):
        return self.seats - len(self._students)

    @property
    def subject(self):
        return self._subject

    @subject.setter
    def subject(self, new_subject:str):
        self._subject = new_subject.title()

    @property
    def students(self) -> list[str]:
        return [name for name, _ in self._students]

    def __str__(self):
        return f'Training "{self.subject}" ({len(self._students)} students)'

    def __repr__(self):
        return f'Training("{self.subject}", {self.duration}, {self.seats})'

    def add_student(self, name:str, company:str=None):
        if len(self._students) >= self.seats:
            raise ValueError('Training full')

        self._students.append((name, datetime.now()))

    @property
    def has_seats_left(self):
        return self.seats - len(self._students) > 0

    def __len__(self):
        return len(self._students)

    def __contains__(self, item):
        return item in [name for name, _, in self._students]
```

```python
t = Training("Python", 5)
t.add_student("Elliot Alderson")
t.add_student("Kevin Mitnik")
```
