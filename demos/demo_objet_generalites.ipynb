## Héritage

```python
class Media:
    def __init__(self, title:str, duration:int):
        self.title = title.title()
        self.duration = duration

class Episode(Media):
    def __init__(self, title:str, number:int, season_number:int, duration:int):
        super().__init__(title, duration)
        self.number = number
        self.season_number = season_number
```

```python
Episode.mro()
```

```python
ep = Episode("My Title", 1, 1, 45)
```

```python
isinstance(ep, Media)
```

## Classes abstraites

```python
from abc import ABC, abstractmethod

class Employee(ABC):
    def __init__(self, first_name, last_name, id):
        self.first_name = first_name
        self.last_name = last_name
        self.id = id

    @abstractmethod
    def calculate_payroll(self):
        pass
```

```python
Employee("john", "Doe", 42)
```

```python
class SalaryEmployee(Employee):
    def __init__(self, first_name, last_name, id, salaray):
        super().__init__(first_name, last_name, id)
        self.monthly_salary = salaray

    def calculate_payroll(self):
        return self.monthly_salary
```

```python
john = SalaryEmployee("John", "Doe", 42, 4000)
```

```python
class Episode:
    __slots__ = ("title", "duration", "number", "season_number")
    def __init__(self, title:str, number:int, season_number:int, duration:int):
        self.title = title
        self.duration = duration
        self.number = number
        self.season_number = season_number
```

```python
ep = Episode("My Title", 1, 2, 45)
```

```python
ep.duration
```

```python
del ep.title
```
