```python
camelot = ["Arthur", "Merlin", "Lancelot", "Robin", "Perceval", "Karadoc"]
```

```python
class Kingdom:
    def __init__(self, knights):
        self.knights = knights

    def __iter__(self):
        self._current_value = 0
        return self

    def __next__(self):
        try:
            value = self.knights[self._current_value]
            self._current_value += 1
            return value
        except IndexError:
            raise StopIteration("End of collection")
```

```python
k = Kingdom(camelot)
```

```python
k.__iter__()
```

```python
for knight in k:
    print(knight)
```
