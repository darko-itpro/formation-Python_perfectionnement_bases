```python
camelot = [("Arthur", "king"), ("Merlin", "wizard"),
           ("Lancelot", "knight"), ("Robin", "knight"),
           ("Perceval", "knight"), ("Karadoc", "knight")]
```

```python
[name for name, _ in camelot]
```

```python
for_quest = [name.upper()
             for name, job in camelot
             if job == "knight"]
```

```python
for i in (name for name, _ in camelot):
    print(i)
```

```python
def for_quest(kingdom:list):
    for name, job in kingdom:
        if job == "knight":
            yield name
    print("Done")
```

```python
for name in for_quest(camelot):
    print(name)
```
