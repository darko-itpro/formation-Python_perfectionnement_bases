# List comprehension

```python
camelot = ["Arthur", "Merlin", "Lancelot", "Robin", "Perceval", "Karadoc"]
```

```python
[len(knight) for knight in camelot]
```

```python
camelot = [("Arthur", "king"), ("Merlin", "wizard"),
           ("Lancelot", "knight"), ("Robin", "knight"),
           ("Perceval", "knight"), ("Karadoc", "knight")]

[name for name, job in camelot if job != "wizard"]
```

# Plusieurs collections

```python
camelot = ["Arthur", "Merlin", "Lancelot", "Robin", "Perceval", "Karadoc"]
```

```python
jobs = ["King", "wizard", "knight", "knight", "knight", "knight", "queen"]
```

```python
for i in zip(camelot, jobs):
    print(i)
```

```python
[(name, job) for name, job in zip(camelot, jobs)]
```

```python
{name:job for name, job in zip(camelot, jobs)}
```

```python
camelot = ["Arthur", "Merlin", "Lancelot", "Robin", "Perceval", "Karadoc"]

start = 0
step = 2

while group := camelot[start:start + step]:
    print(group)
    start += step
```
