```python
camelot = [("Arthur", "king"), ("Merlin", "wizard"),
           ("Lancelot", "knight"), ("Robin", "knight"),
           ("Perceval", "knight"), ("Karadoc", "knight")]
```

Comprehension list

```python
[name for name, _ in camelot]
```

```python
[len(name) for name, job in camelot]
```

```python
[knight
 for knight in camelot
 if knight[1] != "wizard"]
```

```python
[name for name, job in camelot if job != "wizard"]
```

```python
[(name, job) for name, job in camelot if job != "wizard"]
```

```python
names = ["Arthur", "Merlin", "Lancelot", "Robin", "Perceval", "Karadoc"]
jobs = ['king', 'wizard', 'knight', 'knight', 'knight', 'knight']
```

```python
for i in zip(names, jobs):
    print(i)
```

```python
[knight for knight in zip(names, jobs)]
```

```python
{name:job for name, job in zip(names, jobs)}
```
