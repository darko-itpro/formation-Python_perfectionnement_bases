```python
from collections import namedtuple
```

```python
Point = namedtuple("Point", ("x", "y", "z"), defaults=[0,])
```

```python
p = Point(2, 8)
```

```python
p.x
```

```python
p.y
```

```python
p.z
```

```python
p.z = 10
```

```python
from dataclasses import dataclass

@dataclass(frozen=True)
class Episode:
    title: str
    number: int
    season_number: int
    duration: int=None
```

```python
ep = Episode("Gallifrey", 1, 2, 54)
```

```python
ep2 = Episode("Gallifrey", 1, 2, 54)
```

```python
ep == ep2
```
