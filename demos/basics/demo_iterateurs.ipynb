# Les itérateurs

```python
camelot = ["Arthur", "Merlin", "Lancelot", "Robin", "Perceval", "Karadoc"]

class DemoIter:
    def __init__(self, kingdom):
        self.kingdom = kingdom

    def __iter__(self):
        self._current_index = 0
        return self

    def __next__(self):
        try:
            value = self.kingdom[self._current_index]
            self._current_index += 1
            return value
        except IndexError:
            raise StopIteration("End of collection")
```

```python
camiter = DemoIter(camelot)
```

```python
for name in camiter:
    print(name)
```
