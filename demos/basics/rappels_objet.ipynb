```python
from datetime import datetime

class Training:
    def __init__(self, subject:str, seats:int = 12):
        self.subject = subject
        self._students = []
        self.max_seats = seats

    @property
    def available_seats(self):
        return self.max_seats - len(self._students)

    @property
    def subject(self):
        return self._subject

    @subject.setter
    def subject(self, new_name):
        self._subject = new_name.title()

    @property
    def students(self) -> list[str]:
        return [name for name, _ in self._students]

    @students.deleter
    def students(self):
        self._students.clear()


    def __str__(self):
        return f"Training {self.subject} ({len(self._students)} students)"

    def __repr__(self):
        return f"Training('{self.subject}', {self.available_seats})"

    def __eq__(self, other):
        if not isinstance(other, self.__class__):
            return False

        return (self.subject, self.available_seats) == (other.subject, other.available_seats)

    def add_student(self, student:str):
        if len(self._students) >= self.max_seats:
            raise ValueError("Training full")
        self._students.append((student, datetime.now()))

    def __len__(self):
        return len(self._students)

    def __contains__(self, item):
        return item in [name for name, _ in self._students]

    def __getitem__(self, item):
        items = {name:(name, date) for name, date in self._students}
        return items[item]
```

```python
class Training:
    def __init__(self, subject:str, seats:int = 12):
        self.subject = subject
        self._students = []
        self.max_seats = seats

    def add_student(self, student:str):
        if len(self._students) >= self.max_seats:
            raise ValueError("Training full")
        self._students.append((student, datetime.now()))

    @classmethod
    def with_students(cls, students:list[str]):
        training = cls("Python", 5)
        for student in students:
            training.add_student(student)

        return training
```

```python
t = Training.with_students(['Johne Doe', "Kevin Mitnick"])
t._students
```
