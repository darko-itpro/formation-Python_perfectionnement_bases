# Générateurs

```python
camelot = ["Arthur", "Merlin", "Lancelot", "Robin", "Perceval", "Karadoc"]
```

```python
[len(name) for name in camelot]
```

```python
for i in (len(name) for name in camelot):
    print(i)
```

```python
def name_counter(kingdom:list[str]):
    for name in kingdom:
        yield len(name)
    for name in kingdom[::-1]:
        yield len(name)
```

```python
for i in  name_counter(camelot):
    print(i)
```

```python
def repas():
    yield "apero"
    yield "entrée"
    yield "plat"
    yield "dessert"
    yield "café"
```

```python
midi = repas()
```

```python
next(midi)
```

```python
def click():
    count = 0
    while True:
        count += 1
        yield count
```

```python
c = click()
```

```python
next(c)
```
