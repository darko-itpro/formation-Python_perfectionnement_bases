```python
class ContextManager:
    def __init__(self, name):
        self.name = name

    def do_something(self):
        print("Doing something")

    def __enter__(self):
        print("Entering", self.name)
        return self

    def __exit__(self, exc_type, exc_val, exc_tb):
        print("Exiting")
        print(f"{exc_type=}")
        print(f"{exc_val=}")
        print(f"{exc_tb=}")
```

```python
with ContextManager("test") as mycontext:
    mycontext.do_something()
    print("in context")
```

```python
mycm = ContextManager("other_test")
```

```python
with mycm:
    print("in the context")
    mycm.do_something()
    print("still in context")
```

```python
with ContextManager("test"):
    print('in')
    raise ValueError("oups")
```
