# Les décorateurs

```python
def sandwich(garniture):
    def wrapper(*args, **kwargs):
        print('/TTTTTTTT\\')
        price = garniture(*args, **kwargs)
        print('\\________/')
        return price + 2
    
    wrapper.__name__ = garniture.__name__
    wrapper.__doc__ = garniture.__doc__

    return wrapper
```

```python
@sandwich
def parisien():
    """Docstring"""
    print("Jambon")
    print("beurre")
    return 2

@sandwich
def bagnat():
    print("Rillettes thon")
    print("salade")
    return 4

@sandwich
def gyros(salade=True):
    if salade:
        print("salade")
    print('viande')
    return 4
```

```python
print(gyros(False))
```

```python
gyros(salade=False)
```

## Décorateurs paramétrés

```python
def sandwich(hot=False):
    def make_sandwich(garniture):
        def wrapper(*args, **kwargs):
            print('/TTTTTTTT\\')
            price = garniture(*args, **kwargs)
            print('\\________/')
            if hot:
                print("Chauffer 2 minutes")
            return price + 2

        return wrapper

    return make_sandwich
```

```python
@sandwich(hot=True)
def gyros(salade=True):
    if salade:
        print("salade")
    print('viande')
    return 4
```
