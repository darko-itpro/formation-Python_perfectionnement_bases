# Datatypes

## namedtuples

```python
from collections import namedtuple
```

```python
Point = namedtuple("Point", ("x", "y", "z"), defaults=[0,])
```

```python
p = Point(4, 7)
```

```python
p.x
```

```python
p.y
```

```python
p3d = Point(3, 8, 10)
```

```python
p3d.z
```
