# Itérateurs

```python
class DemoIter:
    def __init__(self, kingdom:list):
        self._kingdom = kingdom

    def __iter__(self):
        self._cur_position = 0
        return self

    def __next__(self):
        try:
            element = self._kingdom[self._cur_position].upper()
            self._cur_position += 1
        except IndexError:
            raise StopIteration()

        return element

```

```python
camelot = ["Arthur", "Merlin", "Lancelot", "Robin", "Perceval", "Karadoc"]
```

```python
iterable = DemoIter(camelot)

for knight in iterable:
    print(knight)
```

```python
i = iterable.__iter__()
```
