# Les Contexgt Manager

```python
class DemoCm:
    def __init__(self, name:str):
        self.name = name

    def ping(self):
        print(f"Ping from {self.name}")

    def __enter__(self):
        print("In enter")
        return self

    def __exit__(self, exc_type, exc_val, exc_tb):
        print("In exit")
        print(exc_type)
        print(exc_val)
        print(exc_tb)
        print("End of exit")
        return True
```

```python
with DemoCm("My Cm") as cm2:
    print('In context')
    cm2.ping()
    raise ValueError('oups')
```

```python
cm = DemoCm("My CM")

with cm:
    print("In context")
    cm.ping()
```
