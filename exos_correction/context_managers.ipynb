## Exercice 1

```python
import time

class Timer:
    def __enter__(self):
        self._start = time.process_time()

    def __exit__(self, exc_type, exc_val, exc_tb):
        end = time.process_time()

        print(end - self._start, "secondes se sont écoulées")

with Timer:
    for x in range(1_000_000):
        y = x ** 2
```

# Exercice 2

```python
class IndentContext:
    def __init__(self):
        self._indent = -1

    def print(self, text):
        print(f"{'\t' * self._indent}{text}")

    def __enter__(self):
        self._indent += 1
        return self

    def __exit__(self, exc_type, exc_val, exc_tb):
        self._indent -= 1
```

```python
with IndentContext() as ic:
    ic.print("une ligne")
    ic.print("une autre ligne")
    with ic:
        ic.print("une ligne indentée")
    ic.print("une autre ligne")
```

# Exercice 3

```python
import os
from pathlib import Path

class ChangeDir:
    def __init__(self, target_dir):
        self._targer_dir = Path(target_dir)
        if not self._targer_dir.is_dir():
            pass
        self._from_dir = None

    def __enter__(self):
        if self._from_dir is not None:
            pass
        self._from_dir = Path('.').resolve()
        os.chdir(self._targer_dir)

    def __exit__(self, exc_type, exc_val, exc_tb):
        os.chdir(self._from_dir)
        self._from_dir = None
```

```python
print(os.getcwd())

with ChangeDir('/opt'):
    print(os.getcwd())

print(os.getcwd())
```
