## Exercice 1

```python
import time

def timeit(func):
    def wrapper():
        start = time.process_time()
        func()
        end = time.process_time()
        print(end - start, "secondes se sont écoulées")

    return wrapper

@timeit
def some_func():
    for x in range(1_000_000):
        y = x ** 2

```

```python
some_func()
```

# Exercice 2

```python
def register(func):
    NOTIFICATIONS.append(func)
    return func

NOTIFICATIONS = []

@register()
def notify1():
    print("notification from first")

def notify2():
    print("notification from second")

@register
def notify3():
    print("notification from third")

def send_notifications():
    for notify in NOTIFICATIONS:
        notify()

send_notifications()
```

# Exercice 3

```python
def count_calls(func):
    calls = 0
    def wrapper():
        nonlocal calls
        calls += 1
        func()

    def nbcall():
        return calls

    wrapper.nbcalls = nbcall

    return wrapper

@count_calls
def some_func():
    print("func called")

print(some_func.nbcalls())
some_func()
some_func()
print(some_func.nbcalls())
```

## Exercice 4

```python
import time
from collections import OrderedDict

def cachable(func):
    previous = OrderedDict()

    def inner(*args):
        try:
            value = previous[args]
            previous.move_to_end(args, last=True)
        except KeyError:
            value = func(*args)
            previous[args] = value
            if len(previous) > 5:
                previous.popitem(last=False)

        print(previous)
        return value

    return inner

@cachable
def long_call(value:int):
    time.sleep(2)
    return value**2
```
