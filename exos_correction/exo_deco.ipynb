## Exercice 1

```python
import time

def timeit(func):
    def wrapper():
        start = time.process_time()
        func()
        end = time.process_time()
        print(end - start, "secondes se sont écoulées")

    return wrapper

@timeit
def some_func():
    for x in range(1_000_000):
        y = x ** 2

```

## Exercice 4

```python
import time
from collections import OrderedDict

def cachable(func):
    previous = OrderedDict()

    def inner(*args):
        try:
            value = previous[args]
            previous.move_to_end(args, last=True)
        except KeyError:
            value = func(*args)
            previous[args] = value
            if len(previous) > 5:
                previous.popitem(last=False)

        print(previous)
        return value

    return inner

@cachable
def long_call(value:int):
    time.sleep(2)
    return value**2
```

```python
long_call(2)
```
