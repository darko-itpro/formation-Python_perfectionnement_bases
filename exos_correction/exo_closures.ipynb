# Exo closures

```python
import time
from collections import OrderedDict

def long_call():
    previous = OrderedDict()

    def inner(x):
        try:
            value = previous[x]
            previous.move_to_end(x, last=True)
        except KeyError:
            time.sleep(2)
            value = x**2
            previous[x] = value
            if len(previous) > 5:
                previous.popitem(last=False)

        print(previous)
        return value

    return inner
```

```python
lc = long_call()
```
